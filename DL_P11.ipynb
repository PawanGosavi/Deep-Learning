```python
# Loading the Coco DataSet Model
from imageai.Detection import ObjectDetection
import os

path = '/content/drive/MyDrive/Files/'

detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath(path + "resnet50_coco_best_v2.1.0.h5")
detector.loadModel()
```

```python
# Implementing for Sample Image 1

detections1 = detector.detectObjectsFromImage(input_image = path + "sampleinput1.jpeg", output_image_path = path + "sampleoutput1.jpeg")

for eachObject in detections1:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
```

```python
# Implementing for Sample Image 2

detections2 = detector.detectObjectsFromImage(input_image = path + "sampleinput2.jpeg", output_image_path = path + "sampleoutput2.jpeg")

for eachObject in detections2:
    print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
```
